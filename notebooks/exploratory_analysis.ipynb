## 1. Setup & Environment Configurations

```python
import os
import sys
import pandas as pd
import numpy as np

# Append root folder so we can access our packages
sys.path.append(os.path.abspath('..'))

from ml.preprocessing import DataLoader, DatasetValidator, preprocess_data
from visualization import VisualizerEDA

print("AttriSense AI packages loaded successfully.")
```

## 2. Dataset Loading & High-Level Schema Validation
We use our modular `DataLoader` and `DatasetValidator` to confirm column types, verify boundaries, and audit data quality.

```python
dataset_path = "../datasets/raw/ibm_hr_attrition.csv"

# Initialize loader
loader = DataLoader(dataset_path)
raw_df = loader.load_data()

# Generate metadata profile summary
meta_info = loader.get_basic_info()
print("\n--- Dataset Profile Overview ---")
print(f"Rows: {meta_info['row_count']}, Columns: {meta_info['column_count']}")
print(f"Duplicate Rows: {meta_info['duplicate_rows']}")
print(f"Total Missing Fields: {meta_info['missing_values_total']}")
```

```python
# Execute schema and logical boundary validation
validator = DatasetValidator(target_column="Attrition")
is_valid, errors = validator.validate_schema(raw_df)
quality_report = validator.validate_quality_rules(raw_df)

print("\n--- Schema Integrity Verification ---")
print(f"Is Valid Schema? {is_valid}")
if not is_valid:
    print(f"Validation Errors: {errors}")
print(f"Quality Audit Findings: {quality_report}")
```

## 3. Exploratory Data Analysis & Professional Visualizations
We employ the `VisualizerEDA` suite to output descriptive plots and identify potential structural issues such as target class imbalances or extreme skewness.

```python
# Instantiate the visualizer
viz = VisualizerEDA(output_dir="../visualization/reports")

# 1. Plot Class Distribution on Attrition
viz.plot_target_distribution(raw_df, target_col="Attrition")

# 2. Plot numerical feature distributions
viz.plot_numerical_distributions(raw_df, ["Age", "DailyRate", "DistanceFromHome", "MonthlyIncome"])

# 3. Plot Correlation Matrix heatmap
viz.plot_correlation_heatmap(raw_df, ["Age", "DailyRate", "DistanceFromHome", "MonthlyIncome"])

# 4. Plot comparative Categorical vs Target variables
viz.plot_categorical_vs_target(raw_df, ["BusinessTravel", "Department"])

# 5. Generate Box & Violin plots for outliers
viz.plot_box_and_violin_outliers(raw_df, "MonthlyIncome")

print("All professional charts and plots exported and saved to '../visualization/reports/'.")
```

## 4. Run the Complete, End-to-End Modular Preprocessing Pipeline
To prevent data leakage, the pipeline integrates missing value imputation, outlier handling, feature engineering, encoding, numeric scaling, and class imbalance mitigation in a fully unified fit-transform workflow.

```python
print("Executing full pipeline execution...")
train_df, test_df = preprocess_data(dataset_path, output_dir="../datasets/processed")

print("\n--- Pipeline Output Success ---")
print(f"Processed Training Set Shape: {train_df.shape}")
print(f"Processed Test Set Shape: {test_df.shape}")
print(f"Preprocessed Target Class Balance in Train:\n{train_df['Attrition'].value_counts()}")
```
